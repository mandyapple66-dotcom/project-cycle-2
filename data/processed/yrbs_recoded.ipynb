```python
# ==========================================
# 01_Data_Check.ipynb
# Task: 4.2 Variable Definition and Data Check
# Responsible: Group Member A
# ==========================================

import pandas as pd
import os

# 1. 讀取原始資料 (依照手冊要求，從 data/raw 讀取)
raw_path = '../data/raw/YRBS_2007.csv'

try:
    df = pd.read_csv(raw_path)
    print("✅ 原始資料載入成功")
except FileNotFoundError:
    print("❌ 找不到檔案，請確認 YRBS_2007.csv 放在 data/raw/ 資料夾下")

# 2. 數據檢查與清洗 (Data Check & Cleaning)
# 定義研究變數：憂鬱傾向 (SadOrHopeless) 與 BMI 百分比 (BMIPCT)
target_vars = ['SadOrHopeless', 'BMIPCT']

# 移除目標變數中的缺失值 (NaN)，這是推論統計的重要前提
df_clean = df.dropna(subset=target_vars).copy()

# 3. 執行重編碼 (Recoding Rules)
# 將 SadOrHopeless 從 (1, 2) 轉換為 (1=Success, 0=Failure)
# 1: Yes (感覺憂鬱), 2: No (沒有)
df_clean['is_sad'] = df_clean['SadOrHopeless'].apply(lambda x: 1 if x == 1 else 0)

# 4. 驗證結果 (用於填寫 variable_definitions.md)
print("-" * 30)
print(f"清理後的最終樣本數 (Final Sample Size): {len(df_clean)}")
print("-" * 30)
print("憂鬱傾向變數重編碼統計 (is_sad):")
# 這邊的 1 就是 Success Count
print(df_clean['is_sad'].value_counts()) 

print("\nBMI 百分比統計摘要 (BMIPCT):")
print(df_clean['BMIPCT'].describe())
print("-" * 30)

# 5. 儲存處理後的資料 (存入 data/processed)
os.makedirs('../data/processed', exist_ok=True)
output_path = '../data/processed/yrbs_recoded.csv'
df_clean.to_csv(output_path, index=False)

print(f"🚀 處理後的檔案已儲存至: {output_path}")
print("請記得將此 CSV 檔案上傳到 GitHub 的 data/processed 資料夾！")
```
