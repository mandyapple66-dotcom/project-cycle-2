```python
import pandas as pd

# 讀取檔案
df = pd.read_csv('YRBS_2007.csv')

# 顯示前五筆資料，確認是否讀取成功
df.head()
```

```python
df_sad = df[['SadOrHopeless']].dropna().copy()
df_sad['is_sad'] = df_sad['SadOrHopeless'].apply(lambda x: 1 if x == 1 else 0)
```

```python
df_bmi = df[['BMIPCT']].dropna().copy()
```

```python
print(f"Final sample size for SadOrHopeless: {len(df_sad)}")
print(f"Final sample size for BMIPCT: {len(df_bmi)}")
```

```python
print("\nSuccess Count (SadOrHopeless = 1):")
print(df_sad['is_sad'].value_counts())
```
